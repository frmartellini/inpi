# Sistema GALO de emissão de GRU será desativado dia 10/07

Com a nova modalidade de Guia de Recolhimento da União (GRU) registrada, a partir de 10/07/17, a emissão da guia em lote não mais ocorrerá por meio do sistema GALO. O usuário deverá possuir ferramenta para execução em lote de novo serviço web, que devolverá o arquivo "pdf" da GRU, individualmente. A chamada a este novo serviço será realizada via método POST e protocolo HTTPS. Os parâmetros da chamada serão repassados via JSON, conforme abaixo:

```
{
   "login": string,
   "senha": string,
   "cpfCnpjInpiCliente": string com 11 digitos ou cnpj com 14 digitos,
   "servico": [
      {
         "codigo": int,
         "guiaAnterior": {},
         "valor": float com tres digitos e um ponto,
         "apresentacao": [
            "1",
            "2",
            "3",
            "4"
         ],
         "natureza": {},
         "tipoPeticionamento": [
            "0",
            "1"
         ],
         "numeroProcesso": {},
         "peticaoVinculada": {},
         "total": {},
         "revista": {},
         "objetoPeticao": {}
      }
   ]
}
```

Os parâmetros são descritos a seguir:

1) login - o mesmo utilizado para acesso à geração de GRU, peticionamentos eletrônicos e busca

2) senha - a mesma utilizada para acesso à geração de GRU, peticionamentos eletrônicos e busca

3) cpfCnpjInpiCliente - número do CPF/CNPJ do requisitante ou número INPI, no caso de clientes estrangeiros

4) serviço:

a) codigo - código do serviço

b) guiaAnterior - número da GRU anterior no caso de serviços que exijam o preenchimento do número de GRU paga anteriormente, como complementação de retribuição

c) valor - valor da GRU para serviços sem valor determinado, como complementação de retribuição

d) apresentacao - código numérico da apresentação para o caso de depósito de Marcas, sendo:

          1 - Nominativa
          2 - Figurativa
          3 - Mista
          4 - Tridimensional

e) natureza - código numérico da natureza para os serviços que exigem esta informação, como os depósitos de Marcas, Patentes e DI, sendo:

          i) DI:

                1 - Depósito de Pedidos de Registro de Desenho Industrial (DI)
                2 - Pedidos Divididos

          ii) Patentes:

                1 - Patente de Invenção (PI)
                2 - Certificado de Adição (C)
                3 - Modelo de Utilidade (MU)
                4 - Patente de Invenção (PI) via PCT
                5 - Patente de Invenção (PI) dividido
                6 - Modelo de Utilidade (MU) via PCT
                7 - Modelo de Utilidade (MU) dividido

          iii) Marcas:

                1 - Produto
                2 - Serviço
                3 - Coletiva
                4 - Certificação

 f) tipoPeticionamento - código numérico para os serviços que exigem esta informação, sendo:

                0 - Papel
                1 - Eletrônico

 g) numeroProcesso - número de processo para as petições relacionadas a processos

 h) peticaoVinculada - número de petição para as serviços relacionados a petições anteriores

 i) total - total de itens para serviços que exigem esta informação, como os pedidos de averbação de contrato

 j) revista - número da RPI para serviços que exigem esta informação, como os pedidos de retificação por erro de publicação na RPI

 k) objetoPeticao - código numérico do tipo de objeto da petição para os serviços que exigem esta informação

 Fonte: [INPI](https://www.gov.br/inpi/pt-br/assuntos/noticias/sistema-galo-de-emissao-de-gru-sera-desativado) com adaptações minhas no JSON.
